```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.rc('figure', figsize=(8,6))
mpl.rc('axes', linewidth=2, labelsize=18, labelpad=10)
mpl.rc('xtick', direction='in', top=True, labelsize=14)
mpl.rc('ytick', direction='in', right=True, labelsize=14)
mpl.rc('xtick.major', size=4, width=2)
mpl.rc('ytick.major', size=4, width=2)
mpl.rc('lines', linewidth=2)
mpl.rc('legend', fontsize=13)
mpl.rc('font', family='Serif')
```

```python
lr_data = np.loadtxt('full_sph/exp2/lr_data.txt', delimiter=',')
E_data = np.loadtxt('full_sph/exp2/E_data.txt', delimiter=',')
t, lr2_cd, lr5_cd, lr9_cd, lr2_cr, lr5_cr, lr9_cr = lr_data.transpose()
t, Ek_cd, Ep_cd, Etot_cd, Ek_cr, Ep_cr, Etot_cr = E_data.transpose()
```

```python
fig = plt.figure(figsize=(9,12))
ax1 = fig.add_subplot(211, xlabel='Time [Myr]', ylabel='Lagrange Radius [pc]')
ax1.plot(t, lr9_cd, label='90%')
ax1.plot(t, lr5_cd, label='50%')
ax1.plot(t, lr2_cd, label='20%')
ax1.set_ylim(0, 33)
ax1.legend()

ax2 = fig.add_subplot(212, xlabel='Time [Myr]', ylabel='Normalized Energy')
ax2.plot(t, Ek_cr, c='C0', ls='-', label='Ek, cluster')
ax2.plot(t, Ep_cr, c='C0', ls='-.',label='Ep, cluster')
#ax.plot(t, Etot_cr, c='C0', ls='-', label='Etot, cluster')
ax2.plot(t, -Ek_cd, c='C1', ls='-', label='Ek, cloud')
ax2.plot(t, -Ep_cd, c='C1', ls='-.',label='Ep, cloud')
#ax.plot(t, Etot_cd, c='C1', ls='-', label='Etot, cloud')
ax2.set_ylim(-1.5, 2.35)
ax2.legend()

fig.subplots_adjust(hspace=0.25)
None
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('Time [Myr]')
ax.set_ylabel('Normalized Energy')
ax.plot(t, Ek_cr, c='C0', ls='-', label='Ek, cluster')
ax.plot(t, Ep_cr, c='C0', ls='-.',label='Ep, cluster')
#ax.plot(t, Etot_cr, c='C0', ls='-', label='Etot, cluster')
ax.plot(t, -Ek_cd, c='C1', ls='-', label='Ek, cloud')
ax.plot(t, -Ep_cd, c='C1', ls='-.',label='Ep, cloud')
#ax.plot(t, Etot_cd, c='C1', ls='-', label='Etot, cloud')
ax.set_ylim(-1.5, 2.0)
ax.legend()
None
```
